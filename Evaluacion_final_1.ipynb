```python
#la clase tienda online debe tener los siguientes atributos
#inventario (lista de diccionarios): Un atributo para almacenar los productos en el inventario.
#Cada producto debe ser representado como un diccionario con las siguientes claves:
#'nombre', 'precio', y 'cantidad'.
class TiendaOnline:
    def __init__(self):
        self.inventario  = []
        self.ventas_totales = 0
#agregar producto
    def agregar_producto (self , nombre, precio , cantidad):
        for producto in self.inventario: #recorre el inventario
            if producto["nombre"]==nombre: #para saber si ya existen el o los productos
                producto ["cantidad"] += cantidad
                return
        agregar_nuevo_producto = {"nombre": nombre , 
                                  "precio": f"$ {precio}" , # se hace la funcion precio para que muestre el signo en el precio
                                  "cantidad": cantidad}
        self.inventario.append (agregar_nuevo_producto) #Si no existe el producto agregalo al final
```
